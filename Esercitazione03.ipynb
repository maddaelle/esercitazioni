```python
def POSNEG(x):
  if x < 0:
    return 'il num è neg'
  elif x == 0:
    return 'il num inserito è zero, cambiare num'
  else:
    return 'il num è pos'
```

```python
POSNEG(1)
```

```python
def PARIDISPARI(x):
  if x % 2 == 0:
    return 'il numero è pari'
  else:
    return 'il numero è dispari'
```

```python
x = int(input('inserire numero: '))
print(PARIDISPARI(x))
```
